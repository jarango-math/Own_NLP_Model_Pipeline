## text cleaning
from the initial veiw we can see several steps to be taken for cleaning the text
1. Remove NaN values
2. make everything lowercase
3. remove stop words
4. implement stemming
5. rejoin text

```python
df.isnull().sum()
```

```python
df=df.dropna(subset=['Review Text'])
```

```python
sentiment_features=pd.DataFrame()
sentiment_features['text']=df['Review Text']
sentiment_features['text']=sentiment_features['text'].str.lower()
sentiment_features['rating']=df['Rating']
```

```python
from sklearn.model_selection import train_test_split
Train_set, Test_set = train_test_split(sentiment_features, train_size=.80)
```

```python
from nltk.corpus import stopwords
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")
stop = stopwords.words('english')
```

```python
Train_set['text']=Train_set['text'].str.split()
```

```python
Train_set['text']=Train_set['text'].apply(lambda x: [item for item in x if item not in stop])
Train_set['text']=Train_set['text'].apply(lambda x:[stemmer.stem(item) for item in x])
```

```python
Train_set['text']=Train_set['text'].str.join(" ")
```

## Method 1
Bag of words using CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(Train_set.text)
X_train_counts.shape
count_vect.vocabulary_
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
train = lr.fit(X_train_counts, Train_set['rating'])
print('Training set score:', lr.score(X_train_counts, Train_set['rating']))
```

```python
Test_set['text']=Test_set['text'].str.split()
Test_set['text']=Test_set['text'].apply(lambda x: [item for item in x if item not in stop])
Test_set['text']=Test_set['text'].apply(lambda x:[stemmer.stem(item) for item in x])
Test_set['text']=Test_set['text'].str.join(" ")
```

```python
X_test_counts = count_vect.transform(Test_set.text)
X_test_counts.shape
count_vect.vocabulary_
```

```python
print('Training set score:', lr.score(X_test_counts, Test_set['rating']))
```

## Method 2
tfidf score

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(max_df=0.5, # drop words that occur in more than half the paragraphs
                             min_df=2, # only use words that appear at least twice
                             stop_words='english', 
                             lowercase=True, #convert everything to lower case (since Alice in Wonderland has the HABIT of CAPITALIZING WORDS for EMPHASIS)
                             use_idf=True,#we definitely want to use inverse document frequencies in our weighting
                             norm=u'l2', #Applies a correction factor so that longer paragraphs and shorter paragraphs get treated equally
                             smooth_idf=True #Adds 1 to all document frequencies, as if an extra document existed that used every word once.  Prevents divide-by-zero errors
                            )
```

```python
X_train_tfidf = vectorizer.fit_transform(Train_set.text)
```

```python
X_test_tfidf = vectorizer.transform(Test_set.text)
```

```python
lr_tfidf = LogisticRegression()
train_tfidf = lr_tfidf.fit(X_train_tfidf, Train_set['rating'])
print('Training set score:', lr_tfidf.score(X_train_tfidf, Train_set['rating']))
print('Training set score:', lr_tfidf.score(X_test_tfidf, Test_set['rating']))
```
